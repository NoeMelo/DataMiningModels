```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../Dataset/outliers/censusn.csv",delimiter=";")
data.head()
```

```python
data = data.iloc[:,1:]#eliminamos el ID
data.head()
```

```python
data.columns
```

```python
numericos = ['age', 'final.weight', 'education.num','capital.gain', 'capital.loss', 'hours.per.week']
categoricos = ['employment','education','marital.status', 'job', 'relationship', 'race', 'gender','native.country']
```

```python
fig = plt.subplots(figsize = (15,10))

for i, col in enumerate(numericos):
    plt.subplot(2,3,i+1)
    #data = df[col]
    sns.boxplot(x=data[col], y=data["salary"], data=data)
```

```python
for i, col in enumerate(categoricos):
    sns.catplot(y=col, hue="salary", kind="count",data=data)
```
